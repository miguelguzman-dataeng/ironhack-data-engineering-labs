```python

# Lab 4 — Billing Calculator with Default Arguments

# Step 1: Define the function
def calculate_total(amount, tax_rate=0.05):

    # Step 2: Calculate the tax
    tax = amount * tax_rate

    # Step 3: Calculate the final total
    total = amount + tax

    # Step 4: Send the total back
    return total

# Step 5: Use the default 5% tax
print(f"Total: {calculate_total(100):.2f}")  

# Step 6: Use a custom 12% tax
print(f"Total: {calculate_total(200, 0.12):.2f}")

```

```python
# Lab 5 — Discount Calculator

# Step 1: Define the function
def apply_discount(price, discount_percent):
    """Return the price after applying a percentage discount."""

    # Step 2: Calculate the discounted price
    discounted_price = price * (1 - discount_percent / 100)

    # Step 3: Send the answer back
    return discounted_price

# Step 4: Call the function
result = apply_discount(2000, 6)

# Step 5: Show the returned result
print(result)

# Step 6: Read the function documentation
print(apply_discount.__doc__)

```

```python
# Lab 6 — Normalize Names

# Step 1: Define the cleaning function                 
def normalize_name(name):

    # Step 2: Remove spaces from both ends
    stripped_name = name.strip()

    # Step 3: Break the name into separate words
    words = stripped_name.split()

    # Step 4: Join the words using one space
    single_spaced_name = " ".join(words)

    # Step 5: Convert each word to title case
    normalized_name = single_spaced_name.title()

    # Step 6: Send the cleaned name back
    return normalized_name

# Step 7: Test the first messy name
result_one = normalize_name("pHOebe")
print(result_one)

# Step 8: Test the second messy name
result_two = normalize_name("miguEL")
print(result_two)

```

```python
# Lab 7 — Basic Stats

# Step 1: Define the function
def basic_stats(numbers):

    # Step 2: Count how many numbers exist
    count = len(numbers)   

    # Step 3: Add all numbers
    total = sum(numbers)

    # Step 4: Calculate mean safely
    if count > 0:
        mean = total / count
    else:
        mean = none

    # Step 5: Return all three values
    return count, total, mean

# Step 6: Create sample data
numbers = [10, 20, 30]

# Step 7: Call function and unpack returned values
count_result, total_result, mean_result = basic_stats(numbers)

# Step 8: Print results
print("Count:",count_result )
print("Total:", total_result)
print("Mean:",mean_result )

```

```python


# Lab 8 — Apply Tax to a List

# Step 1: Define the function
def apply_tax(amounts, tax_rate=0.1):

    # Step 2: Create a new empty result list
    taxed_amounts = []

    # Step 3: Process every amount
    for amount in amounts:

        # Step 4: Calculate the taxed amount
        taxed_amount = amount * (1 + tax_rate)

        # Step 5: Add result to the new list
        taxed_amounts.append(taxed_amount)

    # Step 6: Send the new list back
    return taxed_amounts

# Step 7: Original data
amounts = [100, 200, 350]

# Step 8: Call function with 5% tax
result = apply_tax(amounts, 0.05)

# Step 9: Print transformed result
print(result)

# Step 10: Confirm original list did not change
print("Original:", amounts)
```

```python
#Lab 9 Function with Optional Argument and Simple Validation

# Define a function that receives a score.

# max_score is optional and uses 100 when no value is provided.
def process_score(score, max_score=100):

    # Check whether the score is invalid.
    # A score cannot be negative or greater than the maximum score.
    if score < 0 or score > max_score:
        return None

    # Convert the score into a percentage.
    percent_score = (score / max_score) * 100

    # Return the percentage rounded to 1 decimal place.
    return round(percent_score, 1)


# 85 is valid because it is between 0 and 100.
print(process_score(85))

# 150 is invalid because it is greater than the default maximum of 100.
print(process_score(-10))
```

```python

# Lab 15 — Integrated Lab: Sales ETL with Functions + NumPy
import numpy as np


# ==========================================================
# RAW INPUT DATA
# ==========================================================

raw_sales = [
    120.0,
    250.5,
    np.nan,
    -20.0,
    400.0,
    150.0,
    0.0,
    300.0
]


# ==========================================================
# FUNCTION 1.Transform: CLEAN THE SALES DATA
# ==========================================================

def clean_sales(arr):
    
    # 1.Convert to float array
    # dtype=float ensures every value is stored as a decimal number.
    sales_array = np.array(arr, dtype=float)

    # 2.Create a Boolean mask for invalid numbers.
    # Negative numbers and zero become True.
    invalid_mask = sales_array <= 0

    # 3.Replace all invalid values selected by the mask with np.nan.
    # np.nan means "not a number" or missing/invalid numeric data.
    sales_array[invalid_mask] = np.nan

    # np.isnan(sales_array) finds all np.nan values.
    # The ~ symbol reverses the mask:
    # True for valid values and False for np.nan values.
    cleaned_array = sales_array[~np.isnan(sales_array)]

    # Send the cleaned NumPy array back to the caller.
    return cleaned_array


# ==========================================================
# FUNCTION 2.Transform: APPLY DISCOUNT TO LARGE SALES
# ==========================================================

def apply_discount_to_large_sales(
    arr,
    threshold=200,
    discount_rate=0.1
):
    """
    Apply a discount to sales strictly greater than the threshold
    and return a new NumPy array.
    """

    # Create a copy so the original array is not changed.
    discounted_array = arr.copy()

    # Create a Boolean mask.
    # Only values strictly greater than 200 become True
    # when the default threshold is used.
    large_sales_mask = discounted_array > threshold

    # Reduce selected values by the discount rate.
    # With discount_rate = 0.1:
    # 1 - 0.1 = 0.9, so the customer pays 90%.
    discounted_array[large_sales_mask] = (
        discounted_array[large_sales_mask]
        * (1 - discount_rate)
    )

    # Return the new array containing discounted values.
    return discounted_array


# ==========================================================
# FUNCTION 3.LOAD : SUMMARIZE THE SALES
# ==========================================================

def summarize_sales(arr):
    """
    Return a dictionary containing the total, count,
    mean, minimum, and maximum sales values.
    """

    # Create a dictionary containing all summary calculations.
    summary = {
        # Add all sales values together.
        "total": round(float(np.sum(arr)), 2),

        # Count how many values exist in the array.
        "count": int(arr.size),

        # Calculate the average and round it to 2 decimal places.
        "mean": round(float(np.mean(arr)), 2),

        # Find the smallest sales value.
        "min": float(np.min(arr)),

        # Find the largest sales value.
        "max": float(np.max(arr))
    }

    # Return the complete dictionary.
    return summary


# ==========================================================
# # CALL THE FUNCTIONS
# ==========================================================

# STEP 1 — CLEAN
# Remove np.nan, negative numbers, and zero.
cleaned_sales = clean_sales(raw_sales)

# STEP 2 — TRANSFORM
# Apply a 10% discount to values strictly greater than 200.
discounted_sales = apply_discount_to_large_sales(cleaned_sales)

# STEP 3 — SUMMARIZE
# Calculate total, count, mean, minimum, and maximum.
summary = summarize_sales(discounted_sales)


# ==========================================================
# PRINT THE RESULTS
# ==========================================================

# :.2f shows exactly two decimal places.
print(f"Total: {summary['total']:.2f}")

print(f"Count: {summary['count']}")

print(f"Mean: {summary['mean']:.2f}")

print(f"Min: {summary['min']}")

print(f"Max: {summary['max']}")



```
