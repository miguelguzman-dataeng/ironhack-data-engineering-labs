## Lab 1 — Data Lake vs Data Warehouse vs Lakehouse

### A. Finance Team

I would choose a **Data Warehouse**.

The Finance team needs clean, structured, reliable data in a consistent table format for daily reporting. A data warehouse is designed for this type of analytical and business reporting workload.

### B. Data Science Team

I would choose a **Data Lake**.

The Data Science team wants to keep raw structured, semi-structured, and unstructured data such as logs, clickstreams, and images. A data lake is suitable because it can store large amounts of raw data in flexible formats.

## Lab 2 — Schema-on-Read vs Schema-on-Write

The Data Science team benefits from **schema-on-read** because the raw clickstream data can be stored immediately even when some fields are missing or change over time.

The Finance team needs **schema-on-write** because the reporting data must have consistent columns, correct data types, and a predictable structure before it is used.

The main risk of schema-on-read for Finance is that missing fields, inconsistent structures, or incorrect data types could cause reporting errors or break dashboards.

## Lab 4 — The Silver Zone

The Silver layer contains data that has been cleaned, standardized, validated, and transformed into a more reliable form.

A shared Silver dataset prevents different teams from repeatedly cleaning the same raw data in different ways.

Silver is usually not the final dashboard layer because the data may still need business-specific calculations, joins, or aggregations.

## Lab 6 — Separating Compute from Storage

Separating compute from storage means the data is stored independently from the system that processes it.

Different tools can use the same central storage while using their own compute resources.

This gives the company more flexibility because the data is not locked into one processing engine.

# Capstone — Full Data Pipeline Design

## 1. Raw App Event Data

The raw app event data should first be stored in the **Bronze layer**.

The original data should be preserved so it can be audited, debugged, or reprocessed later.

## 2. Bronze to Silver

To move the data into Silver, I would:

- remove duplicate events
- validate required fields
- correct or standardize data types
- standardize column names
- handle missing or invalid values
- standardize timestamps
- validate event types

## 3. Gold Table

I would create a Gold table called:

`fact_daily_app_activity`

Possible columns:

- date
- active_users
- screen_views
- clicks
- purchases
- total_revenue
- conversion_rate

The Gold table would be used by executives and analysts for dashboards and reporting.

## 4. Fixing a Purchase Counting Bug

If a bug is discovered in the transformation logic, I would fix the transformation code and reprocess the original data starting from the Bronze layer.

This is why keeping the original Bronze data is important.

## 5. Architecture Choice

I would choose a **Lakehouse architecture**.

The application produces large amounts of raw and changing data that benefits from flexible data lake storage, while the business also needs reliable structured tables and SQL analytics.

A lakehouse combines both needs in one architecture.
