```python
# Lab 13 — Find top customer

# Step 1: Customer totals
totals = {
    "Asha": 1200.50,
    "Rajan": 3200.00,
    "Leela": 2450.75
}

# Step 2: Create variables that remember the current winner
max_name = " "
max_amount = 0.0 

# Step 3: Loop through every customer and amount
for customer_name, amount in totals.items():

    # Step 4: Check whether current amount is the biggest so far
    if amount > max_amount:

        # Step 5: Remember the new biggest amount and customer
        max_amount = amount
        max_name = customer_name

# Step 6: Print the final winner
print(f"Top customer: {max_name} with {max_amount}")
```

```python
# Lab 14 — While loop simulating balance updates

# Step 1: Starting data
balance = 500.0
transactions = [100.0, -50.0, 0.0, -600.0, 200.0]

# Step 2: Start at the first list position
index = 0  

# Step 3: Remember whether processing stopped early
stopped_early = False

# Step 4: Continue while the index is inside the list
while index < len(transactions):

    # Take the current transaction
    transaction = transactions[index]

    # Move to the next position
    index += 1

    # Skip zero transactions
    if transaction == 0.0:
        continue

    # Apply transaction to balance
    balance += transaction

    # Stop if balance becomes negative
    if balance < 0:
        stopped_early = True
        break

# Step 5: Print status if processing stopped early
if stopped_early:
    print("Processing stopped early due to negative balance.")

# Step 6: Print final balance
print(f"Final balance: {balance}")
```

```python
# Lab 16 — Integrated Lab: Orders mini-ETL

# EXTRACT: raw order data
orders = [
    {"order_id": 1, "customer": "Asha", "amount": 250.0, "product_id": 1001, "status": "delivered"},
    {"order_id": 2, "customer": "Ravi", "amount": 100.0, "product_id": 1002, "status": "pending"},
    {"order_id": 3, "customer": "Asha", "amount": 150.0, "product_id": 1003, "status": "delivered"},
    {"order_id": 4, "customer": "Meera", "amount": 400.0, "product_id": 1001, "status": "delivered"},
    {"order_id": 5, "customer": "Ravi", "amount": 50.0, "product_id": 1002, "status": "cancelled"}
]

# TRANSFORM: create empty result containers
totals_by_customer = {}      
unique_products = set()      
delivered_count = 0

# Process every order
for order in orders:

    # Skip orders that are not delivered
    if order["status"] != "delivered":
        continue

    # Get fields from the delivered order
    customer = order["customer"]
    amount = order["amount"]
    product_id = order["product_id"]

    # Add amount to this customer's running total
    totals_by_customer[customer] = totals_by_customer.get(customer, 0) + amount

    # Add product ID to unique-products set
    unique_products.add(product_id)

    # Count delivered order
    delivered_count += 1

# LOAD: create final nested report
report = {
    "totals_by_customer":totals_by_customer,
    "unique_products":unique_products ,
    "delivered_count":delivered_count
}

# Print report clearly
print("Totals by customer:")

for customer, total in report["totals_by_customer"].items():
    print(f"{customer}: {total}")

print()
print("Unique products sold (delivered):", report["unique_products"])
print("Delivered order count:", report["delivered_count"])
```
