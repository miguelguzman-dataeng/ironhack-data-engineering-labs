```python
# Lab 12: Marks percentage calculator
# 1. Incoming grade data as strings
m1 = "78"
m2 = "85"
m3 = "69"

# 2. Type cast strings to integers
score1 = int(m1)  # Pattern Example!
score2 = int(m2)
score3 = int(m3)

# 3. Do the math
total_earned = score1 + score2 + score3
percentage = (total_earned / 300) * 100

# 4. Print the formatted percentage line
print(f"Percentage: {percentage:.2f}%")

# 5. Smart Rule Fork (if/else preview)
if percentage >= 35:
    print("Result: True")
else:
    print("Result: Fail")
```

```python
# Lab 13: Discount calculation and rounding
# 1. Given data
original_price = 4999.99
discount_percent = 12.5

# 2. Do the calculations using parentheses
discount_amount = original_price * (discount_percent / 100)  # Pattern Example!
final_price = original_price - discount_amount

# 3. Print the formatted prices with 2 decimal places
print(f"Discount: {discount_amount:.2f}")
print(f"Final Price: {final_price:.2f}")
```
